```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import numpy as np
os.chdir('C:\\Users\\guilh\\OneDrive\\Documentos\\Projetos\\DesmatamentoXMultas')
plt.style.use('tableau-colorblind10')
```

```python
with open('multas.pkl','rb') as file:
    multas_dict = pickle.load(file)
desmatamento = pd.read_csv('desmatamento.csv', sep =';')
```

```python
cols = desmatamento.columns
new_cols = ['Year','Area_km2']
col_map = {key:value for key,value in zip(cols,new_cols)}
desmatamento.rename(columns = col_map, inplace = True)
```

```python
#cleaning the fines ammount
for key,value in multas_dict.items():
    new = [float(i.replace('.','').replace(',','.')) for i in value]
    multas_dict[key] = new
```

```python
num_multas = [len(multas) for multas in multas_dict.values()]
sum_multas = [sum(multas)/(10**9) for multas in multas_dict.values()]
#retrieving the index of the years of interest
index = desmatamento.loc[(desmatamento['Year']>2009) & (desmatamento['Year']<2021)].index.tolist()
area = desmatamento.iloc[index,1].values
#cleaning the area values
area = [float(i.replace('.','').replace(',','.')) for i in area ]
years = list(range(2010,2021))
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(2,1,figsize =(14,7))
sns.lineplot(x= years, y= area, color= 'black',linewidth= 3.5,ax= ax[0],label= 'Deforested area in km²')
sns.lineplot(x= years, y= num_multas, color= 'g',linewidth= 3.5,ax= ax[0],label= 'Number of fines')
sns.lineplot(x= years, y= sum_multas,linewidth= 3.5, color= '#dde70c',ax= ax[1])
ax[0].set_yticks([0,3000,6000,9000,12000])
ax[1].set_yticks([1.0,1.5,2.0,2.5,3.0])
ax[1].set_xticks(list(range(2010,2021)))
for i in range(0,2):
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['left'].set_visible(False)
    ax[i].set_facecolor('#f5f5f5')
    ax[i].grid(axis= 'y',color='grey',linestyle= '--',dashes= (10, 5),alpha= 0.35)
    ax[i].grid(color= 'grey',linestyle= '--',dashes= (10, 5),alpha= 0.3)
    ax[i].grid(axis= 'x')
    

ax[0].get_xaxis().set_ticks([])

#ax[0].set_title('Total area deforested in km²',fontsize='large')

ax[1].set_title('Total ammount fined in billions of BRL',fontsize='large')
fig.patch.set_facecolor('#f5f5f5')
fig.suptitle('Deforestation of Brazilian Amazon and enviromental fines (2010-2020)',fontsize='x-large',fontweight="bold")
fig.tight_layout(pad=2.5)
ax[0].legend(loc= 7,facecolor= None,edgecolor= None,framealpha =0)
```
